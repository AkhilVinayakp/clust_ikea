```python
import pandas as pd
import numpy as np
```

```python
columns = ["Kommun_code","Year","Kommun_name","Revenue","Employee","Population","Population_University","Percent_University","Productivity","SalesIndex","Infrast","Border"]
```

```python
len(columns)
```

```python
df = pd.DataFrame(columns=columns)
```

```python
with open("ikea_data.txt", "r") as fp:
    lines = fp.readlines()
    
```

```python
for i in lines:
    l = i.split("\t")
#     print(len(l))
    dic = {}
    for i in range(0,12):
        dic[columns[i]] = l[i]
    df = df.append(dic, ignore_index=True)
```

```python
# converting data types
```

```python
df = pd.read_csv('cleaned_not_conv.csv')
```

```python
df.dtypes
```

```python
# int(pk[0].strip('""'))
```

```python
df["Kommun_code"] = df["Kommun_code"].apply(lambda x: int(x.strip('""')))
```

```python
df["Year"] = df["Year"].apply(lambda x: int(x.strip('""')))
```

```python
df
```

```python
df["Revenue"] = df["Revenue"].apply(lambda x: int(x.strip('""')))
df["Employee"] = df["Employee"].apply(lambda x: int(x.strip('""')))
df["Population"] = df["Population"].apply(lambda x: int(x.strip('""')))
df["Population_University"] = df["Population_University"].apply(lambda x: int(x.strip('""')))
```

```python
df
```

```python
df["Percent_University"] = df["Percent_University"].apply(lambda x: float(x.strip('""')))
df["Productivity"] = df["Productivity"].apply(lambda x: float(x.strip('""')))
df["SalesIndex"] = df["SalesIndex"].apply(lambda x: float(x.strip('""')))
```

```python
df
```

```python
df["Infrast"].unique()
```

```python
df["Infrast"] = df["Infrast"].apply(lambda x: int(x.strip('""')))
```

```python
df
```

```python
df["Border"] = df["Border"].apply(lambda x: int(x.rstrip('\n').strip('""')))
```

```python
df
```

```python
df.dtypes
```

```python
df.to_csv("cleaned_ikea.csv",index=False)
```
